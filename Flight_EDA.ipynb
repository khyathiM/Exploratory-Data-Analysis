```python
print('Length of training set before dropping NA cell',len(train_set))  

train_set=train_set.dropna()   

print('Length of training set after dropping NA cell',len(train_set))  
```

```python
train_set['Journey Day'] = pd.to_datetime(train_set.Date_of_Journey, format='%d/%m/%Y').dt.day
train_set['Journey Day'], train_set['Date_of_Journey']
```

```python
train_set['Journey Month'] = pd.to_datetime(train_set.Date_of_Journey, format='%d/%m/%Y').dt.month
train_set['Journey Month'], train_set['Date_of_Journey']
```

```python
test_set['Journey Day'] = pd.to_datetime(test_set.Date_of_Journey, format='%d/%m/%Y').dt.day
test_set['Journey Month'] = pd.to_datetime(test_set.Date_of_Journey, format='%d/%m/%Y').dt.month
test_set.head()
```

```python
train_set.drop(labels = 'Date_of_Journey', axis = 1, inplace = True)
print('training set after transformation')
display(train_set.head())
```

```python
test_set.drop(labels = 'Date_of_Journey', axis = 1, inplace = True)
print('test set after transformation')
display(test_set.head())
```

```python

duration_train = list(train_set['Duration'])


for i in range(len(duration_train)):
    if len(duration_train[i].split()) != 2: 
        if 'h' in duration_train[i] :
            duration_train[i] = duration_train[i].strip() + ' 0m'
        elif 'm' in duration_train[i] :
            duration_train[i] = '0h {}'.format(duration_train[i].strip()) # output will be 0h 5m
            # duration_train[i] = '0h'+ duration_train[i].strip() output will be 0h5m
            # duration_train[i] = '0h {}'+ duration_train[i].strip() output will be 0h {}5m

duration_train_hours = []
duration_train_min = []  


for i in range(len(duration_train)) :
    duration_train_hours.append(int(duration_train[i].split()[0][:-1])) # from ['8h', '20m'], only 8
    duration_train_min.append(int(duration_train[i].split()[1][:-1])) # 2h 10s then 2:0:10
    
train_set['Duration_hours'] = duration_train_hours
train_set['Duration_minutes'] =duration_train_min

train_set.drop(labels = 'Duration', axis = 1, inplace = True)
```

```python
duration = list(test_set['Duration'])


for i in range(len(duration)):
    if len(duration[i].split()) != 2: 
        if 'h' in duration[i] :
            duration[i] = duration[i].strip() + ' 0m'
        elif 'm' in duration[i] :
            duration[i] = '0h {}'.format(duration[i].strip()) # output will be 0h 5m
            # duration[i] = '0h'+ duration[i].strip() output will be 0h5m
            # duration[i] = '0h {}'+ duration[i].strip() output will be 0h {}5m

duration_hours = []
duration_min = []  


for i in range(len(duration)) :
    duration_hours.append(int(duration[i].split()[0][:-1])) #from ['8h', '20m'], only 8
    duration_min.append(int(duration[i].split()[1][:-1]))#0:49:4 and if 2 hours 10 seconds then it will reflect like 2:0:10
    
test_set['Duration_hours'] = duration_hours
test_set['Duration_minutes'] =duration_min

test_set.drop(labels = 'Duration', axis = 1, inplace = True)
```
